```python
gender_mapping={"Female":1,"Male":2,"Other":3}
df["Gender"]=df["Gender"].map(gender_mapping)
```

```python
df=pd.get_dummies(df,columns=["Occupation"],drop_first=True)
```

```python
df["Smoking"]=df["Smoking"].map({"Yes":1,"No":0})
df["Family History of Anxiety"]=df["Family History of Anxiety"].map({"Yes":1,"No":0})
df["Dizziness"]=df["Dizziness"].map({"Yes":1,"No":0})
df["Medication"]=df["Medication"].map({"Yes":1,"No":0})
df["Recent Major Life Event"]=df["Recent Major Life Event"].map({"Yes":1,"No":0})
```

```python
corr_matrix=df.corr()
```

```python
anxiety_corr=corr_matrix["Anxiety Level (1-10)"].sort_values(ascending=True)
anxiety_corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix)
plt.title("Biểu đồ tương quan giữa mức độ lo âu và các tác nhân khác")
```
