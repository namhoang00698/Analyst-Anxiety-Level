```python
df_groupedhs = df.groupby("Sleep Hours")["Anxiety Level (1-10)"].mean().reset_index()
df_groupedhs["Sleep Hours"]=df_groupedhs["Sleep Hours"].round()
plt.figure(figsize=(10,6))
sns.barplot(x=df_groupedhs["Sleep Hours"],y=df_groupedhs["Anxiety Level (1-10)"],data=df_groupedhs,width=0.6)
plt.xlabel("Số giờ ngủ")
plt.ylabel("Mức độ lo âu (1-10)")
plt.xticks(range(0,10))
plt.show()
```

```python
df_groupedpa=df.groupby("Physical Activity (hrs/week)")["Anxiety Level (1-10)"].mean().reset_index()
df_groupedpa["Physical Activity (hrs/week)"]=df_groupedpa["Physical Activity (hrs/week)"].round()
plt.figure(figsize=(10,6))
sns.barplot(x=df_groupedpa["Physical Activity (hrs/week)"],y=df_groupedpa["Anxiety Level (1-10)"],data=df_groupedpa,width=0.6)
plt.xlabel("Tần suất tập thể dục")
plt.ylabel("Mức độ lo lắng (1-10)")
plt.xticks(range(0,10))
plt.show()
```

```python
selected_column=["Sleep Hours","Physical Activity (hrs/week)","Caffeine Intake (mg/day)","Alcohol Consumption (drinks/week)","Stress Level (1-10)","Heart Rate (bpm)","Breathing Rate (breaths/min)","Sweating Level (1-5)","Therapy Sessions (per month)","Diet Quality (1-10)","Anxiety Level (1-10)"]
corr_matrix=df[selected_column].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True)
plt.title("Heatmap thể hiện mối tương quan giữa mức độ lo lắng và các tác nhân khác")
plt.show()
```

```python
df_groupedci=df.groupby("Caffeine Intake (mg/day)")["Anxiety Level (1-10)"].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=df_groupedci["Anxiety Level (1-10)"],y=df_groupedci["Caffeine Intake (mg/day)"],data=df_groupedci,width=0.6)
plt.xlabel("Mức độ lo âu (1-10)")
plt.ylabel("Lượng caffeine tiêu thụ (mg/day)")
plt.xticks(range(0,10))
plt.show()
```

```python
df_numeric = df.select_dtypes(include=["number"])
```

```python
df_corr = df_numeric.corr()
anxiety_corr = df_corr["Anxiety Level (1-10)"].drop('Anxiety Level (1-10)')
top5factor = anxiety_corr.abs().sort_values(ascending = False).head(5)
print("Top 5 tac nhan anh huong lon den Muc do lo au:")
for factor in top5factor.index:
    print(f"{factor}: {anxiety_corr[factor]:.3f}")
```
